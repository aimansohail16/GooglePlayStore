# Type Casting

```python
data['Price'] = data['Price'].str.replace('$', '')
```

```python
data['Reviews'] = pd.to_numeric(data['Reviews'], errors='coerce')
data['Price'] = pd.to_numeric(data['Price'], errors='coerce')
```

```python
data.dtypes
```

# Missing Values Calculation

```python
data.isnull().sum().sort_values(ascending = False)
```

# Handle Missing Values

```python
for column in data.columns:
    if data[column].dtype == 'object':
        data[column].fillna(data[column].mode()[0], inplace=True)
    elif data[column].dtype == np.number:
        data[column].fillna(data[column].median(), inplace=True)
```

```python
data.isnull().sum().sort_values(ascending = False)
```

# Statistical Summary

```python
data.describe()
```

# Understand the relationships between variables

```python
sns.histplot(data['Rating'].dropna(), bins=20, kde=True)
plt.title("Distribution of App Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Apps")
plt.show()
```

```python
# Remove commas and plus signs
data['Installs'] = data['Installs'].str.replace('[+,]', '', regex=True)

# Keep only rows with numeric installs
data = data[data['Installs'].str.isnumeric()]

# Convert to float
data['Installs'] = data['Installs'].astype(float)
sns.scatterplot(x='Installs', y='Rating', data=data)
plt.xscale('log')
plt.title("Rating vs Installs")
plt.xlabel("Installs (log scale)")
plt.ylabel("Rating")
plt.show()
```

```python
top_categories = data['Category'].value_counts().head(10)
sns.barplot(x=top_categories.values, y=top_categories.index)
plt.title("Top 10 Categories by Number of Apps")
plt.xlabel("Number of Apps")
plt.ylabel("Category")
plt.show()
```

```python
sns.countplot(data=data, x='Type')
plt.title("Free vs Paid Apps")
plt.xlabel("App Type")
plt.ylabel("Count")
plt.show()
```

```python
paid_apps = data[data['Type'] == 'Paid']

sns.histplot(paid_apps['Price'], bins=30, kde=True)
plt.title("Price Distribution of Paid Apps")
plt.xlabel("Price ($)")
plt.ylabel("Number of Apps")
plt.xlim(0, 100)  # limit x-axis to ignore outliers
plt.show()
```

```python
avg_rating = data.groupby('Category')['Rating'].mean().sort_values(ascending=False).head(10)
sns.barplot(x=avg_rating.values, y=avg_rating.index)
plt.title("Top 10 Categories by Average Rating")
plt.xlabel("Average Rating")
plt.ylabel("Category")
plt.show()
```

```python
sns.boxplot(x='Type', y=np.log1p(data['Installs']), data=data)
plt.title("Installs by App Type")
plt.ylabel("Log(Installs + 1)")
plt.show()
```

```python
# Clean and convert 'Size' to MB
def convert_size(size):
    if size.endswith('M'):
        return float(size.replace('M', ''))
    elif size.endswith('k'):
        return float(size.replace('k', '')) / 1024
    else:
        return np.nan

data['Size'] = data['Size'].replace('Varies with device', np.nan)
data['Size'] = data['Size'].dropna().apply(convert_size)

sns.scatterplot(x='Size', y='Rating', data=data)
plt.title("App Size vs Rating")
plt.xlabel("Size (MB)")
plt.ylabel("Rating")
plt.show()
```

```python
data['Last Updated'] = pd.to_datetime(data['Last Updated'])
data['Last Updated Year'] = data['Last Updated'].dt.year

sns.boxplot(x='Last Updated Year', y='Rating', data=data)
plt.title("Rating Trend by Last Updated Year")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
corr = data.select_dtypes(include=['number']).corr()
sns.heatmap(corr, vmax=0.8, annot=True);
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.countplot(y='Content Rating', data=data, order=data['Content Rating'].value_counts().index)
plt.title("Content Rating Distribution")
plt.xlabel("Number of Apps")
plt.ylabel("Content Rating")
plt.show()
```

```python
top_reviewed = data[['App', 'Reviews']].sort_values(by='Reviews', ascending=False).head(10)
sns.barplot(x='Reviews', y='App', data=top_reviewed)
plt.title("Top 10 Most Reviewed Apps")
plt.xlabel("Reviews")
plt.ylabel("App")
plt.show()
```

```python
category_installs = data.groupby('Category')['Installs'].sum().sort_values(ascending=False).head(10)
sns.barplot(x=category_installs.values, y=category_installs.index)
plt.title("Top Categories by Total Installs")
plt.xlabel("Total Installs")
plt.ylabel("Category")
plt.show()
```
